```python
extractor = cv2.xfeatures2d.SIFT_create()

def features(path):
  image = cv2.imread(path)
  gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  (keypoints, descriptors) = extractor.detectAndCompute(image, None)
  return keypoints, descriptors
```

```python
def build_histogram(descriptor_list, cluster_alg):
    histogram = np.zeros(len(cluster_alg.cluster_centers_))
    cluster_result =  cluster_alg.predict(descriptor_list)
    for i in cluster_result:
        histogram[i] += 1.0/len(descriptor_list)
    return histogram
```

Vocal


```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 60)
kmeans.fit(descriptors)

preprocessed_image = []

images = train_img_path + test_img_path
label_image = label_path_train + label_test_img 

for image in images:
    keypoints, descriptors = features(image)
    if descriptors is not None:
        hist = build_histogram(descriptors, kmeans)
        preprocessed_image.append(hist)
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler

preprocessed_image = StandardScaler().fit_transform(preprocessed_image)
```

```python
train_feature_vector = []
kmeans.fit(descriptors)

for image in train_img_path:
  keypoints, descriptors = features(image)
  if descriptors is not None:
    hist = build_histogram(descriptors, kmeans)
    train_feature_vector.append(hist)

train_feature_vector = StandardScaler().fit_transform(train_feature_vector)
```

```python
test_feature_vector = []

for image in test_img_path:
    keypoints, descriptors = features(image)
    kmeans.fit(descriptors)
    if descriptors is not None:
        hist = build_histogram(descriptors, kmeans)
        test_feature_vector.append(hist)

test_feature_vector = StandardScaler().fit_transform(test_feature_vector)
```

Local fearure : HOG( Histogram of Oriented Gradients)




```python
images = train_img_path + test_img_path

print(len(images))

img = cv2.imread(images[50], 0)     #images[0-60]
print('Kch thc nh gc: ', img.shape)
# 1. Khai bo cc tham s
cell_size = (8, 8)  # h x w in pixels
block_size = (2, 2)  # h x w in cells
nbins = 9  # number of orientation bins

# 2. Tnh ton cc tham s truyn vo HOGDescriptor
# winSize: Kch thc ca bc nh c crop  chia ht cho cell size.
winSize = (img.shape[1] // cell_size[1] * cell_size[1], img.shape[0] // cell_size[0] * cell_size[0])
# blockSize: Kch thc ca 1 block
blockSize = (block_size[1] * cell_size[1], block_size[0] * cell_size[0])
# blockStride: S bc di chuyn ca block khi thc hin chun ha histogram bc 3
blockStride = (cell_size[1], cell_size[0])
print('Kch thc bc nh crop theo winSize (pixel): ', winSize)
print('Kch thc ca 1 block (pixel): ', blockSize)
print('Kch thc ca block stride (pixel): ', blockStride)
```

```python
# 3. Tnh ton HOG descriptor
hog = cv2.HOGDescriptor(_winSize=winSize,
                        _blockSize=blockSize,
                        _blockStride=blockStride,
                        _cellSize=cell_size,
                        _nbins=nbins)

# Kch thc ca li  vung.
n_cells = (img.shape[0] // cell_size[0], img.shape[1] // cell_size[1])
print('Kch thc li  vung ( vung): ', n_cells)

# Reshape hog feature
hog_feats = hog.compute(img)\
               .reshape(n_cells[1] - block_size[1] + 1,
                        n_cells[0] - block_size[0] + 1,
                        block_size[0], block_size[1], nbins) \
               .transpose((1, 0, 2, 3, 4))  

print('Kch thc hog feature (h, w, block_size_h, block_size_w, nbins): ', hog_feats.shape)
hog_feats = hog_feats.reshape(hog_feats.shape[0],-3)
print('Kch thc hog feature  c reshape: ', hog_feats.shape)
```

```python
def build_histogram(descriptor_list, cluster_alg):
    histogram = np.zeros(len(cluster_alg.cluster_centers_))
    cluster_result =  cluster_alg.predict(descriptor_list)
    for i in cluster_result:
        histogram[i] += 1.0/len(descriptor_list)
    return histogram
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 33)
kmeans.fit(hog_feats)

preprocessed_image_2 = []

images = train_img_path + test_img_path
label_image = label_path_train + label_test_img 

for image in images:
    keypoints, descriptors = features(image)
    if descriptors is not None:
        hist = build_histogram(hog_feats, kmeans)
        preprocessed_image_2.append(hist)
```

```python
# Standardization
print('Standardization')
from sklearn.preprocessing import StandardScaler

preprocessed_image_2 = StandardScaler().fit_transform(preprocessed_image_2)
```

```python
train_feature_vector_2 = []

for image in train_img_path:
    kmeans.fit(hog_feats)
    if hog_feats is not None:
        hist = build_histogram(hog_feats, kmeans)
        train_feature_vector_2.append(hist)

train_feature_vector_2 = StandardScaler().fit_transform(train_feature_vector_2)

```

```python
test_feature_vector = []

for image in test_img_path:
    kmeans.fit(hog_feats)
    if hog_feats is not None:
        hist = build_histogram(hog_feats, kmeans)
        test_feature_vector.append(hist)

test_feature_vector = StandardScaler().fit_transform(test_feature_vector)
```

```python
model = SVC(kernel='linear', probability=True, random_state=42)
model.fit(train_feature_vector_2,label_path_train)
label_pred_des = model.predict(test_feature_vector)
```

```python
print(f'Accuracy SVM: {np.multiply(metrics.accuracy_score(label_test_img, label_pred_des),1)} ')
print(f"Precision SVM: {np.multiply(round(metrics.precision_score(label_test_img, label_pred_des, pos_label='dog'), 4),1)} ")
print(f"Recall SVM: {np.multiply(round(metrics.recall_score(label_test_img, label_pred_des, pos_label='dog'), 4),1)} ")
print(f"f1_score SVM: {np.multiply(round(metrics.f1_score(label_test_img, label_pred_des, pos_label='dog'), 4),1)} ")
```
